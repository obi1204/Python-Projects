```python
import pandas as pd

stocks = pd.read_csv("all_stocks_5yr.csv")
stocks
```

```python
#covert to datetime to group by date
stocks['date'] = pd.to_datetime(stocks.date)
```

```python
stocks['date']
```

```python
#extracting different parts of data into seperate columns

stocks['year'] = pd.DatetimeIndex(stocks.date).year
stocks['month'] = pd.DatetimeIndex(stocks.date).month
stocks['day'] = pd.DatetimeIndex(stocks.date).day
```

```python
#data for the year 2016
stocks_2016 = stocks[stocks.year == 2016]
stocks_2016
```

```python
stock_2016_high_avg = stocks_2016.high.mean()
stock_2016_high_avg
```

```python
year_means = [stock_2013_high_avg, stock_2014_high_avg, stock_2015_high_avg, stock_2016_high_avg, stock_2017_high_avg, stock_2018_high_avg]
```

```python
years = range(2013, 2019)
```

```python
plt.plot(years, year_means, marker = 'o')
plt.xlabel('Year')
plt.ylabel('Avg. of Amazon Stock High Prices');
```

```python
stock_2013_low_avg = amazon_2013.low.mean()
stock_2014_low_avg = amazon_2014.low.mean()
stock_2015_low_avg = amazon_2015.low.mean()
stock_2016_low_avg = amazon_2016.low.mean()
stock_2017_low_avg = amazon_2017.low.mean()
stock_2018_low_avg = amazon_2018.low.mean()
```

```python
year_volume = [stock_2013_low_avg, stock_2014_low_avg, stock_2015_low_avg, stock_2016_low_avg, stock_2017_low_avg, stock_2018_low_avg]
```

```python
plt.plot(years, year_means, marker = 'o', c ='b')
plt.plot(years, year_volume, marker = 's', c = 'r')
plt.xlabel('Year')
plt.ylabel('Avg. of Amazon Stock High Prices')

plt.title("Amazon Stock: High & Low Prices")
plt.legend(['Highs', 'Lows']);
```
