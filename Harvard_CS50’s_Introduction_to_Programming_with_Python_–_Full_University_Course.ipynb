re library: defines and checks for patterns, can extract and change inputted data

re.search(pattern, string, flag=0):

pattern: pattern you search for

string: actual string you want to search for the pattern

flags: modifies behavior of function

. : any character except a newline

*: 0 or more repetitions

+: 1 or more repetitions

?: 0 or 1 repetition

{m}: m repetitions

{m,n}: range m-n repetitions

```python
#can still be valid without domain, must use + instead of *
import re

email = input("What is you email? ").strip()

if re.search(".*@.*", email): #now there must be something to the left and right of the @
  print("Valid")
else:
  print("Invalid")
```

```python
email = input("What is you email? ").strip()

if re.search(".*@.+", email): #now there must be something to the left and right of the @
  print("Valid")
else:
  print("Invalid")
```

```python
email = input("What is you email? ").strip()

if re.search("..@..", email): #now there must be something to the left and right of the @
  print("Valid")
else:
  print("Invalid")
```

Method used to keep track of input matches pattern

Finite State Machine: starts by reading email from left to right, decides whether to stay in first state or transition to the next state.

In the first state, it iterates over each letter until it reaches @, then iterates over next set of letters

Double cirles means if the the computer finds itself in a second accept state it is a valid email, if it get stuck in the first state it is invalid

```python
#a sentence could still work if it includes @ and .com
email = input("What is you email? ").strip()

if re.search(r".+@.+\.com", email): #\. makes it end on a literal .com, r makes sure python does not execute \ like it would usually ex.\n
  print("Valid")
else:
  print("Invalid")
```

^: mathches the start of the string

$: matches the end of the string just before the newline at the end of the string

```python
email = input("What is you email? ").strip()

if re.search(r"^.+@.+\.com$", email):
  print("Valid")
else:
  print("Invalid")
```

[]: set of characters, one or more characters to specifically look for

[^]: complementing the set, you cannot match these characters

```python
email = input("What is you email? ").strip()

if re.search(r"^[a-zA-Z0-9]+@[^@]+\.com$", email): #stops multiple @ signs to be valid, only allows one
  print("Valid")
else:
  print("Invalid")
```
